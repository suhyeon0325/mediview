```python
import pandas as pd
import numpy as np
```

```python
drugs_data = pd.read_csv('data/전처리데이터/의약품제품코드만.csv')
```

```python
# 연령별, 성별 평균 신장 및 체중 데이터
# 예시 데이터이므로 실제 데이터로 대체 필요
age_groups = ['20대', '30대', '40대', '50대', '60대', '70대', '80대', '90대']
height_weight_data = {
    '남성': {
        '평균신장': [175, 175, 174, 173, 171, 169, 167, 165],
        '평균체중': [70, 78, 80, 81, 79, 76, 73, 70]
    },
    '여성': {
        '평균신장': [162, 162, 161, 160, 158, 156, 154, 152],
        '평균체중': [55, 58, 60, 61, 62, 61, 59, 57]
    }
}
```

```python
# 환자 데이터 생성 함수
def generate_patient_data(patient_id):
    gender = np.random.choice(['남성', '여성'])
    
    # 60대 이상에 더 큰 확률 부여
    age_probabilities = [0.05, 0.05, 0.10, 0.10, 0.20, 0.20, 0.15, 0.15]
    age_group = np.random.choice(age_groups, p=age_probabilities)
    age_index = age_groups.index(age_group)

    height = np.random.normal(height_weight_data[gender]['평균신장'][age_index], height_weight_data[gender]['평균신장'][age_index] * 0.1)
    weight = np.random.normal(height_weight_data[gender]['평균체중'][age_index], height_weight_data[gender]['평균체중'][age_index] * 0.1)

    # 알레르기 상태를 80% 확률로 '무', 20% 확률로 '유'
    allergy = np.random.choice(['무', '유'], p=[0.8, 0.2])

    # 랜덤으로 7개의 제품코드 선택
    selected_drugs = drugs_data.sample(7)['제품코드'].tolist()

    patient_data = {
        '환자 ID': patient_id,
        '성별': gender,
        '나이': age_group,
        '신장': round(height, 2),
        '체중': round(weight, 2),
        '알레르기': allergy
    }

    # 각 약물의 제품코드를 딕셔너리에 추가
    for i, code in enumerate(selected_drugs, 1):
        patient_data[f'약물{i} 제품코드'] = code

    return patient_data

# 여러 환자 데이터 생성 및 저장
num_patients = 100
patients_data = [generate_patient_data(i + 1) for i in range(num_patients)]

# 데이터프레임으로 변환 및 출력
patients_df = pd.DataFrame(patients_data)
patients_df
```

```python
patients_df.to_csv("data/랜덤가상데이터_100.csv")
```
