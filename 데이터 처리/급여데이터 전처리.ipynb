```python
df = pd.read_excel("data/약제급여목록및급여상한금액표_(2024.5.1.)(22,850)_공개용.xlsx")
df.head()
```

```python
# 제품코드가 숫자 형식인것만 필터링
df = df[df['제품코드'].str.isnumeric()]
```

```python
df.head()
```

```python
df['제품명'] = df['제품명'].str.split('_').str[0]
```

```python
df.head()
```

```python
df = df.drop(columns=['연번'])
```
