```python
import pandas as pd
```

```python
df_1 = pd.read_csv("data/건강보험심사평가원_연령별 금기 의약품목록_20230501.csv", encoding='cp949')
df_1.head()
```

```python
df_1['제품명'] = df_1['제품명'].str.split('_').str[0]
```

```python
df_1['연령정보'] = df_1['특정연령'].astype(str) + df_1['특정연령단위-1'].str.extract(r'(\D+)')[0] + " " + df_1['특정연령단위-2'].str.extract(r'(\D+)')[0]
```

```python
df_1 = df_1.drop(columns=['특정연령', '특정연령단위-1', '특정연령단위-2'])
```

```python
df_1['연령정보'] = df_1['연령정보'].str.replace(r'[\(\)]', '', regex=True)
```

```python
df_1.head()
```

```python
df_1.to_csv("data/전처리데이터/연령별_금기_의약품목록.csv")
```
