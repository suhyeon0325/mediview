```python
import pandas as pd
```

```python
df1 = pd.read_excel("data/전처리데이터/현재급여가능의약품.xlsx")
df1.head(2)
```

```python
df2 = pd.read_excel("data/게시_DUR점검대상 비급여의약품 품목리스트(8,232).xlsx")
df2.head(2)
```

```python
df3 = pd.read_csv("data/의약품바코드데이터 (1).csv")
df3.head(2)
```

```python
df3.head(1).T
```

```python
df1 = df1.drop(columns=['Unnamed: 0', '상한금액', '비고', '목록정비전코드'])
df1.head(1)
```

```python
df2 = df2.drop(columns=['연번', '영문성분명 및 함량(단위)', '의약품표준코드' , '식약처품목기준코드'])
df2.head(1)
```

```python
df1.columns = ['투여경로', '분류번호', '주성분코드', '제품코드', '제품명', '업체명', '규격', '단위', '전문일반']
df1.head(1)
```

```python
df2.columns = ['제품코드', '제품명', '주성분코드', '분류번호', '규격', '단위', '업체명', '투여경로', '전문일반']
df2.head(1)
```

```python
df4 = pd.concat([df1, df2])
```

```python
df4.info()
```

```python
# 중복된 제품코드행 제거
df5 = df4.drop_duplicates(subset='제품코드', keep='first')
```

```python
df5.info()
```

```python
df5
```

### 바코드데이터랑 급여/비급여 join

```python
# 제품코드 결측치 제거
df3 = df3.dropna(subset=['제품코드(개정후)', '품목기준코드'])
```

```python
df3.info()
```

```python
data = df5.merge(df3, left_on='제품코드', right_on='제품코드(개정후)', how='inner')
```

```python
data.head(1).T
```

```python
data.info()
```

```python
data = data.drop_duplicates(subset='제품코드(개정후)')
```
